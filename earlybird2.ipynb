```python
from skimage import io
import numpy as np
import os
```

# 前處理

```python
img_list = list()
img_list_np = np.array(img_list)
for dirPath, dirNames, fileNames in os.walk('Aberdeen'):
    for i, f in enumerate(fileNames):
        loadFile = os.path.join(dirPath, f)
        img = io.imread(loadFile)
        img_np = np.array(img)
        img_fla = img_np.flatten()
#         print(img_np.shape)
        img_fla = img_fla.reshape(1,1080000)
#         print(img_fla)
        if i == 0:
            img_list_np = img_fla
        else:
#             print(img_list_np)
            img_list_np = np.concatenate((img_list_np, img_fla), axis=0)
img_list_np = img_list_np.T
print(img_list_np)
```

```python
img_mean = np.mean(img_list_np)
print(img_mean)
print(img_list_np.shape)
```

```python
img_mean_np = np.ones(448200000,)
img_mean_np = img_mean_np.reshape(1080000, 415)
img_mean_np = img_mean_np * img_mean
print(img_mean_np)
X = img_list_np - img_mean_np
# img_list_np = img_list_np.reshape(600,600,3,415)
U, s, V = np.linalg.svd(X, full_matrices=False)
```

```python
print(U.shape)
print('________________________________')
print(s.shape)
print('________________________________')
print(V.shape)
```

```python
sigma = np.diag(s)
print(sigma.shape)
print('________________________________')
print(s)
```

```python
test_img = io.imread('Aberdeen/0.jpg')
test_img_np = np.array(test_img)
test_img_fla = test_img_np.flatten()
print(test_img_fla)
print(test_img_fla.shape)
weight = list()
for column in range(415):
    weight.append(np.dot(test_img_fla, U.T[column]))
# image = np.dot(U, np.dot(sigma, V))
```

```python
weight_np = np.array(weight)
print(weight_np)
eigenfaces = np.zeros(1080000,)
eigenface = np.zeros(1080000,)
for column, w in enumerate(weight_np):
    eigenfaces += w * U.T[column]
#     if column == 9:
#         eigenface = w * U.T[column]
#         print(9)
    if column == 10:
        eigenface = w * U.T[column]
        print(10)

print('_______________________________________________________')
test_img_mean_np = np.ones(1080000,)
test_img_mean_np = test_img_mean_np * img_mean
eigenfaces += test_img_mean_np
print(eigenfaces)

print('_______________________________________________________')
eigenface += test_img_mean_np
print(eigenface)
```

```python
eigenfaces -= np.min(eigenfaces)
eigenfaces /= np.max(eigenfaces)
eigenfaces = (eigenfaces*255).astype(np.uint8)

eigenface -= np.min(eigenface)
eigenface /= np.max(eigenface)
eigenface = (eigenface*255).astype(np.uint8)
```
